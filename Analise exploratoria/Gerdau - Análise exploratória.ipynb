## Método Apply

Esse método é útil para modificar colunas ou criar novas colunas a partir de colunas antigas. Este aplica uma função a todos os valores de uma série.

```python
df['age']
```

```python
df['age'].isnull().sum()
```

```python
#para verificar se o valor é nulo, usa-se o numpy com a função isnan()
np.isnan(df['age'][1305])    #True é nulo
```

### Aplicando função de faixa etária

```python
idade = 20

def calc_faixa_etaria(idade):

    if np.isnan(idade):
      return "desconhecida"
    elif idade <= 15:
      return "0-15"
    elif idade <= 30:
      return "15-30"
    elif idade <= 45:
      return "30-45"
    else:
      return "45+"
```

```python
calc_faixa_etaria(27)
```

```python
# quantos passageiros há em cada faixa etária? (0-15, 15-30, 30-45, 45+) fachado superior, aberto inferior
df['age'].apply(calc_faixa_etaria)  #aplicar a função na coluna -> "apply"
```

```python
df['age_group'] = df['age'].apply(calc_faixa_etaria)
```

```python
df
```

```python
#quantos passageiros há em cada faixa etária?
df['age_group'].value_counts()
```

```python
#qual a distribuição de sexo de cada faixa etaria?
df.groupby(['age_group', 'sex'])[['sex']].count()
```

```python
#qual a proporção de morte e sobrevivência de cada faixa etária?
df.groupby(['age_group'])[['survived']].mean()

# grande parte das crianças sobreviveram!
```

```python
df.groupby(['age_group','sex'])[['survived']].agg(['mean', 'count'])
```

```python
# por classe social
df.groupby(['age_group','sex','pclass'])[['survived']].agg(['mean', 'count'])
```

### Extraindo o nomes, sobrenomes e titulos das colunas

```python
def extrair_sobrenome(s):
  return s.split(',')[0]
```

```python
df['name'].apply(extrair_sobrenome)
```

```python
def extrai_titulo(s):
  return s.split(', ')[1].split('.')[0]
```

```python
df['name'].apply(extrai_titulo)
```

```python
# criando novas colunas e armazenando as separações de nomes
df['surname'] = df['name'].apply(extrair_sobrenome)
```

```python
df['surname']
```

```python
df['title'] = df['name'].apply(extrai_titulo)
```

```python
df['title']
```

### Função Lambda
